### **Shell Commands** ###

**A shell is an interface that executes commands. The shell allows the user to communicate (via commands) with the kernel (a program at the core of a computer's operating system).**

**A shell is a Command Line Interpreter (CLI) and is one form of navigating an operating system. The other form is through a Graphical User Interface or GUI which allows a user to navigate using icons, menus and windows.**

### **sed** ###

**The Linux sed command is most commonly used for substituting text.**

```python
%%bash

cat sed_example.txt
```

```python
%%bash

cat sed_example.txt | sed 's/D/d/'
```

```python
%%bash

cat sed_example.txt | sed 's/D/d/g'
```

**if 'D' appears at the beginning of a line (^D).**

```python
%%bash

cat sed_example.txt | sed 's/^D/d/g'
```

** "D$ matches the character 'D' at the end of a line ($ signifies the end of a line).** 

```python
%%bash

cat sed_example.txt | sed 's/D$/d/g'
```

```python
%%bash

cat sed_example.txt | sed 's/[A-Z]/\L&/g'
```

### **awk** ###

```python
%%bash

cat awk_example.txt | awk {'print'}
```

```python
%%bash

cat awk_example.txt | awk {'print $1 $2'}
```

```python
%%bash

cat awk_example.txt | awk {'print $1 "\t" $2'}
```

```python
%%bash

cat awk_example.txt | awk {'print $1 "," $2'}
```

```python
%%bash

cat awk_example.txt | awk '$2 >=5 {print}'
```

```python
%%bash

cat awk_example.txt | awk '$2 >=5 && $1 == "a" {print}'
```

```python
%%bash

cat awk_example.txt | awk '$2 >=5 && $1 == "a" {print $3+5}'
```

#### **awk NF** ####

```python
%%bash

cat awk_example.txt | awk '$NF<=3 {print}'
```

#### **awk NR** ####

```python
%%bash

cat awk_example.txt | awk '$NR<=3 {print}'
```
